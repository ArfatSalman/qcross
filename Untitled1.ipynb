# Crashes in source program

```python
crashes_in_srcs = ['qiskit.qiskit_source.exception', 'cirq.cirq_source.exception', 'pyquil.pyquil_source.exception']
def print_crash_in_source_program():
    for src in crashes_in_srcs:
        if src in df.columns:
            col_len = len(df[['prog_id', src]].dropna())
            print( f"For SRC: {src}, exceptions = {col_len}")
            print(f"percent = {col_len / total_rows * 100}")
        else:
            print( f"For SRC: {src}, exceptions = 0")
        
print_crash_in_source_program()
df[['prog_id', crashes_in_srcs[2]]].dropna()
```

# Crashes in followup program

```python
qiskit_followups = ['qiskit.qiskit_followup.exception']
crashes_in_followups = ['cirq.cirq_followup.exception', 'pyquil.pyquil_followup.exception']
def print_crash_in_followup():
    l = 0
    for s in qiskit_followups:
        l += len(df[['prog_id', s]].dropna())
    print( f"For Qiskit followups, exceptions = {l}")
    print(f"percent = {l / total_rows * 100}")

    for src in crashes_in_followups:
        col_len = len(df[['prog_id', src]].dropna())
        print( f"For followup: {src}, exceptions = {col_len}")
        print(f"percent = {col_len / total_rows * 100}")
        
print_crash_in_followup()
```

# Sources and Followup Comparison

```python
from itertools import combinations
from qcross.start import detect_divergence

all_sources = all_df[['qiskit.qiskit_source.result', 'cirq.cirq_source.result','pyquil.pyquil_source.result']].dropna().values.tolist()

print(f'tested - {len(all_sources)}')
# Sources
all_sources_divergence = 0
for result in all_sources:
    for a, b in combinations(result, 2):
        res = detect_divergence(a, b)
        if res['p-value'] < 0.05:
            all_sources_divergence += 1


print(all_sources_divergence)

# Followups   

all_followups = all_df[['qiskit.qiskit_followup.result', 'cirq.cirq_followup.result','pyquil.pyquil_followup.result']].dropna().values.tolist()
print(f'tested - {len(all_followups)}')
all_followups_divergence = 0
for result in all_followups:
    for a, b in combinations(result, 2):
        if isinstance(a, list) or isinstance(b, list):
            continue
        res = detect_divergence(a, b)
        if res['p-value'] < 0.05:
            all_followups_divergence += 1

print(all_followups_divergence)
```
